```
def bayes(a, b_given_a, b_given_not_a):

    """
    Calculates P(A | B) using bayes theorem.

    Args:
        a: P(A)
        b_given_a: P(B | A)
        b_given_not_a: P(B | ~A)

    Returns:
        P(A | B)
    """

    #p(A|B)=p(B}A).P(A)/P(B)
    #P(A | B) = P(B | A) * P(A) / [P(A)*P(B|A) + P(A~)*P(B|A~)]
    return (b_given_a*a)/(b_given_a*a + (1-a)*(b_given_not_a))
```

```
bayes(.0001,1,.01)
```

#PROBLEM 4

Choose any two of the variables from 3. Draw 1,000 samples, taking the mean of 100 observations each time. Plot the frequencies of the sample means. What distribution does this appear to follow?

```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
from scipy import stats
from numpy.random import seed
%matplotlib inline
```

#NORMAL

```
norm_means = [np.random.normal(1, 4, 100).mean() for _ in range(1000)]
```

```
plt.hist(norm_means)
plt.title("SAMPLE MEANS FROM NORMAL DISTRIBUTION")
plt.show()

print(stats.describe(norm_means))
```

#BINOMIAL

```
binomial_means = [np.random.binomial(100, .7,1).mean() for _ in range(1000)]
```

```
plt.hist(binomial_means)
plt.title("SAMPLE MEANS FROM BINOMIAL DISTRIBUTION")
plt.show()

print(stats.describe(norm_means))
```
